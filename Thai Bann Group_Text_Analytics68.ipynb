#### Inspect data in JSON format (either reading using pretty print or Pandas data frame)

````python
# เปิดไฟล์ JSON ด้วย encoding เป็น UTF-8 และอ่านข้อมูลทั้งหมดเป็น string
with open(file_path, 'r', encoding='utf-8') as f:
    raw_text = f.read()  # raw_text เป็น string ของเนื้อหาไฟล์ทั้งหมด

# แปลง string JSON เป็น Python object (dict หรือ list)
outer = json.loads(raw_text)  # รอบแรก parse

# ตรวจสอบว่า object ที่ได้ยังเป็น string หรือไม่
if isinstance(outer, str):
    outer = json.loads(outer)  # ถ้าใช่ parse อีกครั้ง

# แปลง dict เป็น DataFrame ของ pandas
# orient='index' หมายถึง key ของ dict จะกลายเป็น index ของ DataFrame
df = pd.DataFrame.from_dict(outer, orient='index')

# เลือกเฉพาะคอลัมน์ที่ต้องการและแสดงผลแค่ 5 แถวแรก
display(df[['tweet_id', 'username', 'name', 'profile_picture', 'replies',
            'retweets', 'likes', 'is_retweet', 'posted_time', 'content',
            'hashtags', 'mentions', 'images', 'videos', 'tweet_url']].head())
````

#### Report: จงตอบคำถามต่อไปนี้
* The total number of tweets collected in the file

สามารถใช้ `np.argsort(<list>)[:5]` ซึ่งจะแสดงผล indices ของการเรียงค่าจากน้อยสุดไปมากสุด จำนวน 5 ค่า และ `np.argsort(<list>)[::-1][:5]` จะแสดงผล indices ของการเรียงค่าจากมากไปน้อยจำนวน 5 ค่า (https://numpy.org/doc/stable/reference/generated/numpy.argsort.html)

````python
# 1️ จำนวนทวีตทั้งหมดที่เก็บอยู่ใน DataFrame
total_tweets = len(df)
print("1. The total number of tweets collected in the file")
print("Total number of tweets:", total_tweets)
print('\n-----------------------------------------------------------------------------------------------------------------------------------------------------------')

# 2️ แสดงเนื้อหาของทวีตแรก
print("2. The content of the first tweet")
first_tweet_content = df.iloc[0]["content"]  # ใช้ .iloc[0] เพื่อเข้าถึงแถวแรก
print("\nContent of the first tweet:")
print(first_tweet_content)
print('\n-----------------------------------------------------------------------------------------------------------------------------------------------------------')

# 3️ ตรวจสอบว่าทวีตแรกมี hashtags หรือ mentions หรือไม่
print("3. Is the first tweet contained any hashtags or mentions?")
first_tweet_hashtags = df.iloc[0]["hashtags"]
first_tweet_mentions = df.iloc[0]["mentions"]
print("\nFirst tweet contains hashtags:", bool(first_tweet_hashtags))  # True ถ้ามี hashtag
print("First tweet contains mentions:", bool(first_tweet_mentions))    # True ถ้ามี mention
print('\n-----------------------------------------------------------------------------------------------------------------------------------------------------------')

# 4️ เก็บ hashtags ทั้งหมดจากคอลัมน์ 'hashtags'
print("4. Collect all hashtags from the 'hashtags' column")
all_hashtags = [hashtag for sublist in df['hashtags'] for hashtag in sublist]  # Flatten list ของ hashtags
# กรองเฉพาะ hashtags ที่มีคำว่า "Thailand" (ไม่สนใจตัวพิมพ์ใหญ่/เล็ก)
thailand_hashtags = [hashtag for hashtag in all_hashtags if 'thailand' in hashtag.lower()]
print("All Thailand-related hashtags:", thailand_hashtags)
print('\n-----------------------------------------------------------------------------------------------------------------------------------------------------------')

# 5️ หาทวีตที่มีความนิยมมากที่สุด (จำนวน retweets + replies + likes สูงสุด)
print("5. Find the most popular tweets in your collection of tweets, i.e. the tweets with the largest number of retweet/replies/likes counts")
popularity_score = df['retweets'] + df['replies'] + df['likes']  # สร้าง score ความนิยม
most_popular_index = np.argmax(popularity_score)  # หา index ของทวีตที่มี score สูงสุด
most_popular_tweet = df.iloc[most_popular_index]  # ดึงข้อมูลทวีตนั้น
print("Most popular tweet:")
print("Username:", most_popular_tweet['username'])
print("Content:", most_popular_tweet['content'])
print("Replies:", most_popular_tweet['replies'])
print("Retweets:", most_popular_tweet['retweets'])
print("Likes:", most_popular_tweet['likes'])
print('\n-----------------------------------------------------------------------------------------------------------------------------------------------------------')

# 6️ แสดง 5 ทวีตที่มีความนิยมสูงสุด
print("6. Display the top 5 tweets that are the most popular among your collection")
df['popularity'] = df['replies'] + df['retweets'] + df['likes']  # สร้างคอลัมน์ popularity
top_5_indices = df['popularity'].nlargest(5).index  # หา index ของ top 5 ทวีตที่มี popularity สูงสุด
print("Top 5 most popular tweets:")
print(df.loc[top_5_indices, ['content', 'replies', 'retweets', 'likes', 'popularity']])
print('\n-----------------------------------------------------------------------------------------------------------------------------------------------------------')

# 7️ สร้าง Word Cloud จากคำในเนื้อหาทวีต
print("7. Create a word cloud of words in the contents; however, we note that this word cloud is not a good representation of Thailand as the data size is small")
all_content = " ".join(df['content'])  # รวมเนื้อหาทวีตทั้งหมดเป็น string เดียว

# สร้าง WordCloud object
stopwords = set(STOPWORDS)
stopwords.update(["Thailand", "thailand"])  # เพิ่ม "Thailand" เป็น stopword เพราะอยู่ในทุกทวีต
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(all_content)

# แสดง Word Cloud
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')  # ใช้ bilinear เพื่อให้ภาพเรียบ
plt.axis("off")  # ซ่อนแกน x,y
plt.show()
print('\n-----------------------------------------------------------------------------------------------------------------------------------------------------------')
````

อ่านข้อมูล review

````python
# We only use the first 100,000 reviews
n = 100000
reviews = []
with open(myDirectory+reviewFilePath) as fl:
    for i, line in enumerate(fl):
        reviews.append(json.loads(line))
        if i+1 >= n:
            break
reviews = pd.DataFrame(reviews)
reviews.head()
````

อ่านข้อมูล user

````python
users = []
with open(myDirectory+userFilePath) as fl:
    for i, line in enumerate(fl):
        users.append(json.loads(line))
users = pd.DataFrame(users)
users.head(2)
````

### รวมข้อมูล Reviews, Businesses, และ Users และสำรวจตัวแปรในข้อมูล

มีธุรกิจกี่แบบ และธุรกิจในแต่ละแบบมีจำนวนเท่าไหร่

* ลองแยก `categories` ออกเป็น category ย่อย แล้วสามารถนับหรือใช้ฟังก์ชัน Counter `from collections import Counter` ในการนับจำนวนหรือความถี่ของธุรกิจแต่ละแบบ \
ตัวอย่าง https://www.geeksforgeeks.org/counters-in-python-set-1/

````python
# ----------------- Your code here -----------------
# สมมติว่า df คือ DataFrame ที่รวมข้อมูลแล้ว และมีคอลัมน์ชื่อ 'categories'

# สร้างลิสต์เพื่อเก็บ category ทั้งหมด
all_categories = []
for categories in df['categories'].dropna():   # ใช้ .dropna() เพื่อข้ามค่า NaN (ข้อมูลหาย)
    for category in categories.split(', '):    # แยก string ตาม comma แล้ววนลูปทีละ category
        all_categories.append(category)        # เก็บลง list all_categories

# ใช้ Counter เพื่อนับจำนวนครั้งที่แต่ละ category ปรากฏ
category_counts = Counter(all_categories)

# หาจำนวน category ที่ไม่ซ้ำกันทั้งหมด
unique_categories_count = len(set(all_categories))
print(f"Total number of unique categories: {unique_categories_count}")

# แสดง Top 10 category ที่พบบ่อยที่สุด
print("Top 10 most frequent business categories:")
for category, count in category_counts.most_common():  # .most_common(10) เลือกแค่ 10 อันดับแรก
    print(f"{category}: {count}")

# ทางเลือก: สร้าง DataFrame จาก Counter เพื่อดูง่ายขึ้น
category_df = pd.DataFrame.from_dict(category_counts, orient='index', columns=['Count'])
````

#### วิเคราะห์ประเภทของธุรกิจที่พบมากที่สุด

````python
# ----------------- Your analysis code here -----------------
# สมมติว่า df เป็น DataFrame ที่รวมข้อมูลแล้ว และมีคอลัมน์ 'categories'

# สร้าง list เพื่อเก็บหมวดหมู่ (categories) ทั้งหมด
all_categories = []
for categories in df['categories'].dropna():  # .dropna() ใช้ตัดค่าที่เป็น NaN ออก
    for category in categories.split(', '):   # แยก string ตาม comma และ space
        all_categories.append(category)       # เก็บ category แต่ละตัวลง list

# ใช้ Counter เพื่อนับความถี่ของแต่ละ category
category_counts = Counter(all_categories)

# หาหมวดหมู่ 10 อันดับแรกที่พบบ่อยที่สุด
top_10_categories = category_counts.most_common(10)
top_10_categories_df = pd.DataFrame(top_10_categories, columns=['Category', 'Count'])

# แสดงผลในรูปตาราง
print("Top 10 most popular business categories based on review count:")
print(top_10_categories_df)

# ทำ Visualization ด้วย bar chart
plt.figure(figsize=(5, 3))
sns.barplot(
    x='Count',
    y='Category',
    data=top_10_categories_df,
    palette='viridis'   # ใช้โทนสีแบบ viridis
)

# ตกแต่งกราฟ
plt.title('Most Popular Business Categories', fontsize=14, color='darkblue')
plt.xlabel('Number of Businesses', fontsize=12)
plt.ylabel('Business Category', fontsize=12)
plt.tight_layout()
plt.show()
````

#### หาหมวดหมู่ธุรกิจที่มีจำนวนรีวิวสูงและมีคะแนนดาวเฉลี่ยสูง

````python
# ----------------- Your analysis code here -----------------
# สมมติว่า df มีคอลัมน์ 'categories' (หมวดหมู่ธุรกิจ) และ 'stars_review' (คะแนนรีวิวเป็นตัวเลข)

# สร้าง list เพื่อเก็บคู่ข้อมูล (category + star rating)
category_stars = []
for index, row in df.iterrows():                      # วนลูปทีละแถวใน DataFrame
    if pd.notna(row['categories']):                   # ถ้า categories ไม่ใช่ NaN
        for category in row['categories'].split(', '):# แยก categories ที่คั่นด้วย ", "
            category_stars.append({
                'category': category,
                'stars': row['stars_review']          # เก็บค่าเฉลี่ยดาวของรีวิวนั้น
            })

# แปลงเป็น DataFrame ใหม่
category_stars_df = pd.DataFrame(category_stars)

# คำนวณสถิติสำหรับแต่ละ category
# mean = ค่าเฉลี่ยดาว, count = จำนวนรีวิวทั้งหมด
category_stats = category_stars_df.groupby('category')['stars'].agg(['mean', 'count']).reset_index()
category_stats.columns = ['category', 'average_stars', 'review_count']

# กำหนดเกณฑ์ในการคัดเลือก
review_count_threshold = category_stats['review_count'].quantile(0.75)  # เลือกเฉพาะกลุ่ม Top 25% ของจำนวนรีวิว
average_stars_threshold = 4.0  # ค่าเฉลี่ยดาวต้อง ≥ 4.0

# เลือก categories ที่ทั้งรีวิวเยอะ (>= threshold) และดาวเฉลี่ยสูง (>= 4.0)
popular_high_rated_categories = category_stats[
    (category_stats['review_count'] >= review_count_threshold) &
    (category_stats['average_stars'] >= average_stars_threshold)
]

# เรียงลำดับ โดยเอาค่าเฉลี่ยดาวมากสุดก่อน และตามด้วยจำนวนรีวิว
popular_high_rated_categories = popular_high_rated_categories.sort_values(
    by=['average_stars', 'review_count'], ascending=[False, False]
)

# แสดงผลลัพธ์
print("Business categories with high review count and high average star ratings:")
print(popular_high_rated_categories)
````

## Problem 3: Retrieving Data from Internet

ให้แต่ละกลุ่มคิดหัวข้อที่สนใจ และทำการดึงข้อมูลเกี่ยวกับหัวข้อที่สนใจผ่าน url ด้วยฟังก์ชัน `urllib.request.urlopen` จากแหล่งข้อมูลออนไลน์อย่างน้อย 10 แห่ง รวบรวมข้อมูลให้เป็นหนึ่งเดียว

````python
import urllib.request
from bs4 import BeautifulSoup
# รายการ URL เว็บไซต์ข่าว/เทคโนโลยีที่เข้าถึงง่าย
urls = [
    'https://thethaiger.com/hot-news/technology',
    'https://www.nationthailand.com/category/business/tech',
    'https://www.bangkokpost.com/life/tech',
    'https://www.techinasia.com/tag/thailand',
    'https://www.sanook.com/it/',
    'https://www.thaiware.com/',
    'https://droidsans.com/',
    'https://www.techxcite.com/',
    'https://www.beartai.com/tech',
    'https://digitechasean.com/'
]

# กำหนด User-Agent ให้เหมือนเบราว์เซอร์จริง
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/140.0.0.0 Safari/537.36'
}

# สร้าง list เก็บข้อมูล
data_list = []

for url in urls:
    try:
        # สร้าง request พร้อม header
        req = urllib.request.Request(url, headers=headers)

        # เปิด URL และอ่าน HTML
        with urllib.request.urlopen(req) as response:
            html = response.read().decode('utf-8', errors='ignore')  # decode แบบ ignore errors

            # ใช้ BeautifulSoup แปลง HTML เป็น object
            soup = BeautifulSoup(html, 'html.parser')

            # ดึงข้อความทั้งหมดในหน้าเว็บ (แยกด้วย space) และตัด strip whitespace
            # เอาเพียง 1000 ตัวอักษรแรกเพื่อเก็บตัวอย่าง
            text = soup.get_text(separator=' ', strip=True)[:1000]

            # เก็บ URL และข้อความลง list
            data_list.append({'url': url, 'text': text})

            print(f"ดึงข้อมูลจาก {url} เรียบร้อย")
    except Exception as e:
        print(f"ไม่สามารถดึงข้อมูลจาก {url}: {e}")
````

เตรียมข้อมูลและทำความสะอาดข้อมูลด้วยการทำ lower case, tokenization, removing stopwords (as well as punctuations, unrelated symbols, etc.), stemming, lemmatization เป็นต้น

````python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
import re

# ดาวน์โหลดข้อมูล NLTK ที่จำเป็น
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# สร้างตัว stemmer และ lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))  # โหลด stopwords ภาษาอังกฤษ

# ----------------- ฟังก์ชันทำความสะอาดข้อความ -----------------
def clean_and_lemmatize_text(text):
    # ตรวจสอบว่า input เป็น string หรือไม่
    if not isinstance(text, str):
        return ""

    # แปลงเป็น lowercase
    text = text.lower()

    # ลบตัวอักษรที่ไม่ใช่ a-z และ whitespace (รวม punctuation และตัวเลข)
    text = re.sub(r'[^a-z\s]', '', text)

    # แยกคำ (tokenize) โดยใช้ split()
    tokens = text.split()

    # ลบ stopwords, ทำ stemming และ lemmatization
    cleaned_tokens = [
        lemmatizer.lemmatize(stemmer.stem(word))
        for word in tokens
        if word not in stop_words
    ]

    # รวมคำกลับเป็น string เดียว
    return ' '.join(cleaned_tokens)

# ----------------- ประยุกต์ฟังก์ชันกับ data_list -----------------
for item in data_list:
    item['cleaned_text'] = clean_and_lemmatize_text(item['text'])

# แสดงข้อความดั้งเดิมและข้อความที่ทำความสะอาดแล้ว (แค่ 200 ตัวอักษรแรก)
for i in range(min(5, len(data_list))):
    print(f"Original Text from {data_list[i]['url']}:")
    print(data_list[i]['text'][:200] + '...')  # แสดงแค่ 200 ตัวอักษรแรก
    print(f"Cleaned Text:")
    print(data_list[i]['cleaned_text'][:200] + '...')
    print("-" * 50)
````

Text Mining: ทำการวิเคราะห์ข้อความที่รวบรวมมาได้ พร้อมทั้งแสดงการนำเสนอที่สร้างสรรค์

````python
#----------------------------------------------
# Your code starts here
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# รวมข้อความที่ทำความสะอาดแล้วจากทุกเว็บไซต์เป็น string เดียว
all_cleaned_text = " ".join([item['cleaned_text'] for item in data_list])

# สร้าง Word Cloud
wordcloud = WordCloud(
    width=800,                 # ความกว้างของภาพ
    height=400,                # ความสูงของภาพ
    background_color='white'   # พื้นหลังสีขาว
).generate(all_cleaned_text)

# แสดง Word Cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')  # ใช้ bilinear interpolation ให้ภาพเรียบเนียน
plt.axis('off')                                   # ไม่แสดงแกน X/Y
plt.title('Word Cloud of Collected Text Data', fontsize=16, color='darkblue')
plt.show()
````

````python
# ----------------- Word Frequency Analysis -----------------

# แยกคำจากข้อความทั้งหมด
words = all_cleaned_text.split()

# นับความถี่ของแต่ละคำ
word_counts = Counter(words)

# ดึง 20 คำที่พบบ่อยที่สุด
top_n_words = word_counts.most_common(20)

# แปลงเป็น DataFrame เพื่อสะดวกต่อการ plot
top_words_df = pd.DataFrame(top_n_words, columns=['Word', 'Frequency'])

# ----------------- Visualization -----------------
plt.figure(figsize=(10, 6))
sns.barplot(
    x='Frequency',   # แกน X = จำนวนครั้งที่ปรากฏ
    y='Word',        # แกน Y = คำ
    data=top_words_df,
    palette='viridis'  # ใช้โทนสี Viridis
)
plt.title('Top 20 Most Frequent Words', fontsize=16, color='darkblue')
plt.xlabel('Frequency', fontsize=12)
plt.ylabel('Word', fontsize=12)
plt.show()

# แสดงตารางคำที่พบบ่อยที่สุดและจำนวนครั้ง
print("\nTop 20 most frequent words and their frequencies:")
print(top_words_df)
````

````python
# ----------------- Bigram Analysis -----------------
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
from nltk.tokenize import word_tokenize
from nltk.util import bigrams
import nltk
import pandas as pd

# ดาวน์โหลดข้อมูล NLTK ที่จำเป็น
nltk.download('punkt')
nltk.download('punkt_tab')  # สำหรับ tokenizer ที่รองรับ tab

# ----------------- สร้าง Bigrams -----------------
# Assuming 'all_cleaned_text' ถูกสร้างจากขั้นตอนก่อนหน้า
tokens = word_tokenize(all_cleaned_text)  # แยกคำเป็น tokens
bigram_list = list(bigrams(tokens))       # สร้าง bigrams จาก tokens

# นับความถี่ของแต่ละ bigram
bigram_counts = Counter(bigram_list)

# เลือก Top N bigrams
N = 15  # สามารถปรับจำนวนได้
top_n_bigrams = bigram_counts.most_common(N)

# แปลงข้อมูลเป็น DataFrame สำหรับการ plot
top_bigrams_df = pd.DataFrame(top_n_bigrams, columns=['Bigram', 'Frequency'])
# รวม tuple ของคำเป็น string เพื่อให้ง่ายต่อการ plot
top_bigrams_df['Bigram_str'] = top_bigrams_df['Bigram'].apply(lambda x: ' '.join(x))

# ----------------- Visualization -----------------
plt.figure(figsize=(10, 6))
sns.barplot(
    x='Frequency',
    y='Bigram_str',
    data=top_bigrams_df,
    palette='viridis'
)
plt.title(f'Top {N} Most Frequent Bigrams', fontsize=16, color='darkblue')
plt.xlabel('Frequency', fontsize=12)
plt.ylabel('Bigram', fontsize=12)
plt.show()

# แสดงตาราง Top N bigrams
print(f"\nTop {N} most frequent bigrams and their frequencies:")
print(top_bigrams_df[['Bigram_str', 'Frequency']])
````
